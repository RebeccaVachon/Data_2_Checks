```python
import pandas as pd
import numpy as np

data = pd.read_csv('https://www.irs.gov/pub/irs-soi/eo_ky.csv')

# displaying the top 10 rows of the newly created comparsion_df DataFrame to ensure it was loaded correctly
data.head(10)
```

```python
# dropping unneccessary columns
data = data.drop(['ICO', 'ORGANIZATION', "SUBSECTION", "CLASSIFICATION", "PF_FILING_REQ_CD","FILING_REQ_CD", "ASSET_CD", "INCOME_CD"], axis=1)


#changing the Ruling Date column to a date
data['RULING'] = pd.to_datetime(data['RULING'], format='%Y%m', errors='coerce')

# changing NAN values in three columns to zeros
Columns = ["ASSET_AMT", "INCOME_AMT", "REVENUE_AMT"]
for x in Columns:
    data[x] = data[x].fillna(0)

# displaying the top 10 rows again to make sure data has been cleaned correctly
data.head(10)
```
